```python
import matplotlib.pyplot as plt 
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
#convert data columns to in64 

columns = ["United States","Russia"]
for column in columns:
    country = np.array(data[column])
    new_country = []
    for i in country: 
            i = i.replace(",","")
            new_country.append(i)
    np_country = np.array(new_country).astype(int)
    data[column]= np.array(np_country)
print (data["United States"])
print (data["Russia"])
print (data.info())
```

```python
X = data["Year"]
y = data[['United States', 'Russia', 'United Kingdom', 'France', 'China',
       'Israel', 'India', 'Pakistan', 'North Korea']]
```

```python
from matplotlib.pyplot import figure

fig, ax = plt.subplots()
fig = plt.figure(figsize=(8, 6))
# figure(figsize=(20, 10), dpi=80)
for country in y:
    ax.plot(X, data[country], label=country)
    
#    plt.plot(X, data[country], label=country)
    ax.legend(loc='upper left')
start, end = ax.get_ylim()
ax.figure.set_size_inches(20, 8)
ax.yaxis.set_ticks(np.arange(0, end,4250))
plt.show()
```

```python
fig, axs = plt.subplots(3,3)
axs[0,0].scatter(X, data["United States"])
axs[0,0].set_title("United State")
axs[0,1].scatter(X, data["Russia"], color="orange")
axs[0,1].set_title("Russia")
axs[0,2].scatter(X, data["United Kingdom"], color="green")
axs[0,2].set_title("United Kingdom")
axs[1,0].scatter(X, data["France"], color="red")
axs[1,0].set_title("France")
axs[1,1].scatter(X, data["China"], color="purple")
axs[1,1].set_title("China")
axs[1,2].scatter(X, data["Israel"], color="brown")
axs[1,2].set_title("Israel")
axs[2,0].scatter(X, data["India"], color="pink")
axs[2,0].set_title("India")
axs[2,1].scatter(X, data["Pakistan"], color="gray")
axs[2,1].set_title("Pakistan")
axs[2,2].scatter(X, data["North Korea"], color="lime")
axs[2,2].set_title("North Korea")


fig.figure.set_size_inches(20, 15)
plt.show()
```

```python
US_max = np.array(data.loc[data['United States'] == data["United States"].max(), 'Year'])
US_current = np.array(data.loc[data['Year'] == 2022, 'United States'])

print ("The United States reached its maximum stockpile of nuclear warheads,",data["United States"].max(),"in",US_max[0],".\nSince then, the US has reduced its supply to", US_current[0])
```

```python
Rus_max = np.array(data.loc[data['Russia'] == data["Russia"].max(), 'Year'])
Rus_current = np.array(data.loc[data['Year'] == 2022, 'Russia'])

print ("Russia reached its maximum stockpile of nuclear warheads,",data["Russia"].max(),"in",Rus_max[0],".\nSince then, the US has reduced its supply to", Rus_current[0])
```
